```python
import seaborn as sns
import matplotlib.pyplot as plt
a = sns.load_dataset('iris')
a
sns.lineplot(x="sepal_length", y= "sepal_width", data=a,)
sns.set_style("dark")
plt.title("Iris dataset")
plt.figure(figsize=(5,7))
plt.show()
```

# Adding title

```python
a= sns.load_dataset("Iris")
a
sns.pyplot(x="septal_length",y="septal_width", data=a)
plt.title("flower")
plt.show()
```

# Adding limits

```python
a = sns.load_dataset("Iris")
a
sns.lineplot(x="sepal_length",y="sepal_width", data=a)
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# _02-bar_plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

a= sns.load_dataset("Iris")
a
sns.barplot(x="species",y="sepal_width", data=a)
plt.title("flower")
plt.show()
```

```python
###order import seaborn as sns
import matplotlib.pyplot as plt

a= sns.load_dataset("Iris")
a
sns.barplot(x="sepal_length", y="species", hue="sepal_width", data=a )
plt.title("flower")
plt.show()
```

```python
### adding colour in plots
import seaborn as sns
import matplotlib.pyplot as plt

a= sns.load_dataset("Iris")
a
sns.boxplot(x="sepal_length",y="sepal_width", hue="species", data=a , color="red")
plt.title("flower")
plt.show()
```

```python
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

kashti= sns.load_dataset("titanic")
kashti
sns.barplot(x="class",y="fare", hue="sex", data=kashti)
plt.title("flower")
plt.show()
```

```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

kashti= sns.load_dataset("titanic")
kashti
sns.barplot(x="class",y="fare", hue="sex", data=kashti, saturation=1, estimator=median)
plt.title("flower")
plt.show()
```
